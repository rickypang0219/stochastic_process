```python
# 1D Random Walker
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# random walker function 



def random_walk(N, p):
    """
    N = number of steps 
    p = probability of winning
    """
    X = np.random.choice([-1,1], N, p=[p, 1-p])
    return np.concatenate((np.array([0]), X), axis=0) # Random walker starts 0 
    
```

```python
X_i = random_walk(100,0.5)
X_i
```

```python
W = np.cumsum(X_i)
```

```python
plt.title(r" Simple Random Walk with p = 0.5")
for i in range(5): 
    X_i = random_walk(100,0.5)
    plt.plot( np.cumsum(X_i))

plt.plot(np.sqrt(np.arange(0,100)),"--",linewidth=3, color='black')
plt.plot(-np.sqrt(np.arange(0,100)),"--",linewidth=3,color='black')
plt.xlabel("Number of steps")
plt.ylabel("Random Walk")

plt.savefig("random_walk.svg")
```
